### Loading our graphs from question 1.6

```python
with open("../Pickle/EO_16.file", "rb") as f:
    EO = pickle.load(f)
with open("../Pickle/EC_16.file", "rb") as f:
    EC = pickle.load(f)

# EO: Eyes-opened case
dtf_A_EO = EO.bin_adj_matrix_dtf
dtf_graph_EO = EO.dtf_graph
dtf_matrix_EO = EO.dtf_matrix

# EC: Eyes-closed case
dtf_A_EC = EC.bin_adj_matrix_dtf
dtf_graph_EC = EC.dtf_graph
dtf_matrix_EC = EC.dtf_matrix

#Reading in the channel locations
locations = {re.sub('[^\w]',"",x[1][1]):(x[1]['x'], x[1]['y']) for x in pd.read_csv("../../Data/channel_locations.txt", 
                                                                                    sep = "\s+").iterrows()}
```
